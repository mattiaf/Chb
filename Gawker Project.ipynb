## 1. Reading data

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import scipy
from nltk.corpus import stopwords
import numpy

# Read data into a pandas Dataframe
colnames = ['user_id_hash','relative_path','first_visit_ts','seconds_of_engaged_time']
data_orig = pd.read_csv('gawker-clean.txt', delimiter=' ', names=colnames)
data_orig = data_orig[data_orig.relative_path  != '/'] # removing relative_path == homepage 
```

## 1. How many times have articles been viewed?

Half of the articles in the database have been seen only once. (they were probably articles that originally appeared on the website earlier than the February - July window), and only 10% of them have been seen more than ~50 times.

For the readability of our final graph, we restrict the analysis to the n most seen articles (parameter ntop)


```python
# Count visits to each article
count_visits=data_orig.groupby('relative_path').count()['user_id_hash']

# Plot distribution of visits
plt.rcParams['figure.figsize'] = (6.0, 6.0)
plt.hist(scipy.log10(count_visits))
plt.xlabel('log10 Visits')
plt.ylabel('N of articles')

# A measure of inequality
deciles=numpy.percentile(count_visits, range(0,100,10))
print 'deciles'
print deciles
top10=numpy.percentile(count_visits, range(0,100,1))[90:100]
print '90% to 99% percentiles'
print top10
print 'most viewed article has ', max(count_visits), ' views'


# filtering on most seen articles
ntop = 50 # this parameter can be changed
most_seen=count_visits.order(ascending=False)[0:ntop] # select n most seen articles
data = data_orig[data_orig['relative_path'].isin(most_seen.index)].copy() # filter data
data.index=data.user_id_hash 
```

## 2. Better labels: From urls to significative words.

In our graph, we don't want nodes to be labeled as long urls, such as */heres-the-first-clip-from-farrah-abrahams-sex-tape-493130311*.

I write a small function that erases eiphens and numbers, and drops stopwords (the, from, ...)

```python
stop = stopwords.words('english') # read dictionary of stopwords


# this function transforms a url in something more readable - highlighting significative words.
def make_readable(string):
    string = ''.join([i for i in string if not i.isdigit()]) # erases numbers
    string = string.replace('-', ' ') # replaces - and / with spaces
    string = string.replace('/', ' ')
    string = ' '.join([i for i in string.split() if i not in stop]) # keeps only words not in the stopwords list
    return string

# do the transformation on each url
data.relative_path = data.relative_path.apply(lambda x: make_readable(x))
```

## 3. Creating the graph

Here I create a graph with networkx as follows:  
3.1 Make a list of articles (unique paths in the relative_path column)  
3.2 For each article I create a node. We assign as size of the node the total time spent on the article  
3.3 I create edges by finding all articles j read by users that have also read article i.  
3.4 I assign the weight of edges between nodes j and i as the number of users that read both articles i and j



```python
# list of articles
articles=data.relative_path.unique()

# create empty graph
Graphker=nx.Graph()

# For each article:
for article_i in articles:
    
    # CREATE A NODE. its size will the the total time spent on that article
    engaged_time_mean = data[data.relative_path == article_i]['seconds_of_engaged_time'].sum() # total time spent on that article
    Graphker.add_node(article_i,size=engaged_time_mean) # name = name of article, size = see above
    
    # look for all the users that read that article
    users_article_i=data[data.relative_path==article_i]['user_id_hash'].values

    # count how many users that have read article i, read also article j
    data.ix[users_article_i].groupby('relative_path').count()['user_id_hash']

    # find all connected nodes. two articles i,j are connected if at least 1 user read both i and j
    connected_nodes=data.ix[users_article_i].groupby('relative_path').count()['user_id_hash']
    connected_nodes=connected_nodes.drop(article_i) # removes the edge to itself

    # CREATE EDGES
    # the weight of the edge connecting node i to node j is equal to to how many users read both articles
    for j in range(0,len(connected_nodes.index)):
            Graphker.add_edge(article_i,connected_nodes.index[j],weight=connected_nodes.values[j])

```

## 4. Visualize articles

As a sanity check, I plot the graph using the networkx routine.  
Radius of nodes are proportional to (square root of) total time spent by all users on the articles  
Thickness of lines between nodes i and j is proportional to number of users that read both i and j  

### Farrah Abraham?
Three articles mention the 'reality show personality, author, musician, pornographic actress, and stripper' Farrah Abraham (definition from wikipedia).
People that read one of the three articles are very likely to read also the other two. In fact the nodes of articles mentioning Farrah Abraham in the title are very close to each other

```python
# Area of circles proportional to total time spent on the website
size_circles=[d['size'] for (n,d) in Graphker.nodes(data=True)]
perc_99=numpy.percentile(size_circles, range(1,100,1))[90]

size_circles=[scipy.sqrt(i) for i in size_circles]

# Sizes of edges is proportional to how many users have read both articles i and j
size_edges=[d['weight'] for (u,v,d) in Graphker.edges(data=True)]
size_edges = [float(i)/max(size_edges) * 25 for i in size_edges] # the normalization can be improved...

# positions for all nodes
pos=nx.spring_layout(Graphker,dim=2,k=2) 

# Draw edges
nx.draw_networkx_edges(Graphker,pos,alpha=1,width=size_edges, edge_color='steelblue')

# Draw nodes
nx.draw_networkx_nodes(Graphker,pos,alpha=0.9,node_size=size_circles, node_color='grey')

# Draw Labels
# I draw label only for nodes related to Farrah Abraham
labels = {}    
for (n,d) in Graphker.nodes(data=True):
    test= ('farrah' in n) or ('farah' in n) #does the node contain the word far(r)ah?
    if test == True:
        labels[n] = n
    else:
        labels[n] = ''
nx.draw_networkx_labels(Graphker,pos,labels,font_size=12,font_family='sans-serif', font_color='red')

# Display
plt.rcParams['figure.figsize'] = (20.0, 20.0)
plt.axis('off')
plt.show()
```

## 5. Exporting to JSON , Visualizing in D3

The Python visualization is neat, but abit compressed, and does not allow easily to explore trends.

I take inspiration from the D3 visualization "Force-Directed Graph" by Micheal Bostock http://bl.ocks.org/mbostock/4062045 .  
This visualization is effectively a physics simulation in which:  
a. nodes have an intrinsic repulsive force  
b. edges are the force that tie together nodes.  

I augment the visualization by Bostock with:  
a. area of dots proportional to time spent by the users  
b. thickness on lines proportional to number users in common between the nodes  
c. interactive labeling: passing the mouse over a node will tell you the name of the article  


The next lines create a JSON file containing information on the nodes, modeled after miserables.json at http://bl.ocks.org/mbostock/4062045 

In the example JSON nodes are implicitely indexed 0....n , and edges are defined using the indexes of nodes. I reproduce this by using a dictionary that maps node names into a [0,....,n] index, and using that dictionary in the creation of edges

```python
#opens file
f = open("test.json","w")
f.write("{\n")
f.write('   "nodes":[')

# list of edges and nodes
edges=Graphker.edges()
nodes=Graphker.nodes()

# map nodes to index
dic = {nodes[i]:i for i in range(0,len(nodes))}

# create node objects
for (n,d) in Graphker.nodes(data=True):
    f.write('     {"name":"')  # attribute name
    f.write(n)

    f.write('","time":') # attribute time (spent by users on that article)
    f.write(str(d['size']))

    f.write(',"group":1}') # attribute group (can be used in the future to categorize articles)
    
    if n != nodes[len(nodes)-1]:   # no comma in the last node 
        f.write(',')
    f.write('\n')
    
f.write('  ],\n')
f.write('  "links":[\n')

# create edge (links) objects
for (u,v,d) in Graphker.edges(data=True):
    f.write('     {"source":') #attribute source
    f.write(str(dic[u])) # use dictionary (gets name of article, gives index of nodes)

    f.write(',"target":') # attirbute target
    f.write(str(dic[v]) ) # use dictionary  (gets name of article, gives index of nodes)

    f.write(',"value":') # attribute value (thickness of line = users in common between source and )
    f.write(str(float(d['weight'])/20.))
    f.write('}')
    if (u,v) != edges[len(edges)-1]: # no comma in the last edge
        f.write(',')
    f.write('\n')
    
f.write('  ]\n')
f.write('}')
      
f.close()

```
